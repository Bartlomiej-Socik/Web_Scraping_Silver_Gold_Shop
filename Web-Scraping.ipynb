```python
#Web Scraping Silver Gold Shop
from bs4 import BeautifulSoup
from requests import get
from datetime import date
from openpyxl import load_workbook

today = date.today()

def save_to_excel_Silver(x1,x2,x3,x4):
    sheet = wb['Silver']
    sheet.append((x1, x2, x3, x4))
    
def save_to_excel_Gold(x1,x2,x3,x4):
    sheet = wb['Gold']
    sheet.append((x1, x2, x3, x4))
    
def price_repair(price):
    return float(price.replace(',','.').replace('zł','').replace(' / szt. brutto','').replace(' ','').replace('\xa0',''))

wb = load_workbook('Coins.xlsx')


shop = 'Metal_Market'
URL_1 = 'https://www.metalmarket.eu/pl/menu/srebrne-monety-851.html?filter_traits%5B1%5D=367&filter_traits%5B510%5D=481'
bs_1 = BeautifulSoup(get(URL_1).content)
for offer in bs_1.find_all('div',class_='product'):
    try:
        title = offer.find('a',class_='product__name').get_text()
        price = price_repair(offer.find('strong',class_='price').get_text())
        save_to_excel_Silver(today,shop,title,price)
    except Exception:
        pass
URL_1 = 'https://www.metalmarket.eu/pl/bestsellers/bestseller.html?filter_traits%5B1%5D=277&filter_producer=&filter_traits%5B510%5D=481'
bs_1 = BeautifulSoup(get(URL_1).content)
for offer in bs_1.find_all('div',class_='product'):
    try:
        title = offer.find('a',class_='product__name').get_text()
        price = price_repair(offer.find('strong',class_='price').get_text())
        save_to_excel_Gold(today,shop,title,price)
    except Exception:
        pass

shop = 'Metale_Lokacyjne'
URL_2 = 'https://www.metalelokacyjne.pl/zha_m_Produkty_Monety_srebrne-502.html?filter_traits%5B510%5D=481&filter_stock24availability=cena+od+najni%C5%BCszej'
bs_2 = BeautifulSoup(get(URL_2).content)
for offer in bs_2.find_all('div',class_='product_wrapper'):
    try:
        title = offer.find('a',class_='product-name').get_text()
        price = price_repair(offer.find('span',class_='price').get_text())
        save_to_excel_Silver(today,shop,title,price)
    except Exception:
        pass
URL_2 = 'https://www.metalelokacyjne.pl/zha_m_Produkty_Monety_zlote-546.html?filter_traits%5B1%5D=&filter_producer=&filter_traits%5B510%5D=481&filter_traits%5B69%5D=&filter_stock24availability='
bs_2 = BeautifulSoup(get(URL_2).content)
for offer in bs_2.find_all('div',class_='product_wrapper'):
    try:
        title = offer.find('a',class_='product-name').get_text()
        price = price_repair(offer.find('span',class_='price').get_text())
        save_to_excel_Gold(today,shop,title,price)
    except Exception:
        pass

shop = 'Mennica_Skarbowa'
URL_3 = 'https://www.mennicaskarbowa.pl/pol_m_Srebro_Srebrne-monety-151.html'
bs_3 = BeautifulSoup(get(URL_3).content)
for offer in bs_3.find_all('div',class_='product_wrapper'):
    try:
        title = offer.find('a',class_='product-name').get_text()
        price = price_repair(offer.find('span',class_='price').get_text())
        save_to_excel_Silver(today,shop,title,price)
    except Exception:
        pass
URL_3 = 'https://www.mennicaskarbowa.pl/pol_m_Zloto_Zlote-monety-149.html?filter_pricerange=&filter_traits%5B1339672261%5D=&filter_traits%5B1339671853%5D=1339672254&filter_producer=&filter_traits%5B1339670904%5D='
bs_3 = BeautifulSoup(get(URL_3).content)
for offer in bs_3.find_all('div',class_='product_wrapper'):
    try:
        title = offer.find('a',class_='product-name').get_text()
        price = price_repair(offer.find('span',class_='price').get_text())
        save_to_excel_Gold(today,shop,title,price)
    except Exception:
        pass


shop = 'Szlachetne_Inwestycje'
URL_4 = 'https://sklep.szlachetneinwestycje.pl/srebro/srebrne-monety-bulionowe/'
bs_4 = BeautifulSoup(get(URL_4).content)

for offer in bs_4.find_all('li',class_='product'):
    try:
        title = offer.find('p',class_='woocommerce-loop-product__title').get_text()
        price = price_repair(offer.find('span',class_='woocommerce-Price-amount').get_text())
        save_to_excel_Silver(today,shop,title,price)
    except Exception:
        pass
URL_4 = 'https://sklep.szlachetneinwestycje.pl/zloto/zlote-monety-bulionowe/'
bs_4 = BeautifulSoup(get(URL_4).content)
for offer in bs_4.find_all('li',class_='product'):
    try:
        title = offer.find('p',class_='woocommerce-loop-product__title').get_text()
        price = price_repair(offer.find('span',class_='woocommerce-Price-amount').get_text())
        save_to_excel_Gold(today,shop,title,price)
    except Exception:
        pass
    

shop = 'Mennica_Gdańska'
URL_5 = 'https://mennica-gdanska.pl/pl/43-monety-srebrne?q=Pa%C5%84stwo-Australia-Austria-Kanada-Republika+Po%C5%82udniowej+Afryki-USA-Wielka+Brytania/Waga-31%2C1+g&order=product.price.asc'
bs_5 = BeautifulSoup(get(URL_5).content)

for offer in bs_5.find_all('article',class_='js-product-miniature'):
    try:
        title = offer.find('a',class_='product_name').get_text() + offer.find('div',class_='product-desc').get_text()
        price = price_repair(offer.find('span',class_='price').get_text())
        save_to_excel_Silver(today,shop,title,price)
    except Exception:
        pass
URL_5 = 'https://mennica-gdanska.pl/pl/45-monety-zlote'
bs_5 = BeautifulSoup(get(URL_5).content)
for offer in bs_5.find_all('article',class_='js-product-miniature'):
    try:
        title = offer.find('a',class_='product_name').get_text() + offer.find('div',class_='product-desc').get_text()
        price = price_repair(offer.find('span',class_='price').get_text())
        save_to_excel_Gold(today,shop,title,price)
    except Exception:
        pass

wb.save('Coins.xlsx')
```
